### About the Report

This is a quick map displaying the homeless students in the twin cities. The designation was determined by the Minnesota Department of Education, and datasource can be obtained [here](https://public.education.mn.gov/MDEAnalytics/DataTopic.jsp?TOPICID=2). I used Google API to get the respective coordinates for each school. The purpose of this report is to quickly spread awareness of Homeless youth/families before the Winter holiday season. The code is below; and the corresponding data table.

```python
#Run Code
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline 
import folium

#load data
HL= pd.read_csv('Data/TCHomeless.csv')
#HL['Address']= HL['Address'].replace({"L'O":"LO"})
#HL= HL.query('Num!=527')


def plot_Homeless():
    # generate a new map
    FG_map = folium.Map(location=[44.988326, -93.162926], zoom_start=12,tiles="OpenStreetMap")
    A=HL.query('Homeless_Count>=0 and Num not in (463,527,462)')

    print('This map displays 2019 Homeless Student Count > 16 from October 1st 2019 Enrollment')
    for index, row in A.iterrows(): 
        popup_text = "School: {}<br> Address: {}<br> Homeless_Count: {}<br> Homeless%: {}<br> Total Enrollment: {}"
        popup_text = popup_text.format(row["School"],row["Address"], row['Homeless_Count'],row['Homeless%'],row['Tot_Enroll'] )
        folium.CircleMarker(location=(row["Latitude"],row["Longitude"]),
                            radius=row['Homeless_Count'],
                            color="#E37222",
                            popup=popup_text,
                            fill=True).add_to(FG_map)

    return FG_map

def plot_Homeless2():
    # generate a new map
    FG_map = folium.Map(location=[44.988326, -93.162926], zoom_start=12,tiles="OpenStreetMap")
    A=HL.query('Homeless_Count>4')

    print('This map displays 2019 Homeless Student Count > 4 from October 1st 2019 Enrollment')
    for index, row in A.iterrows(): 
        #popup_text = "School: {}<br> Address: {}<br> Homeless_Count: {}<br> Homeless%: {}<br> Total Enrollment: {}"
        #popup_text = popup_text.format(row["School"],row["Address"], row['Homeless_Count'],row['Homeless%'],row['Tot_Enroll'] )
        folium.CircleMarker(location=(row["Latitude"],row["Longitude"]),
                            radius=row['Homeless_Count'],
                            color="#E37222",
                            #popup=popup_text,
                            fill=True).add_to(FG_map)

    return FG_map
```
