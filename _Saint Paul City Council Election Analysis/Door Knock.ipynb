## Ward 1 Door Knock Map

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline 
import requests # library to handle requests
import folium
import seaborn as sns
import scipy.stats as stats #used to get correlation coefficient

b= pd.read_csv('data/W1_print.csv')


def district(Precinct):
    if Precinct==540:
        return 1
    elif Precinct==550:
        return 2
    elif Precinct==560:
        return 3
    elif Precinct==570:
        return 4
    elif Precinct==580:
        return 5
    elif Precinct==590:
        return 6       
    elif Precinct==600:
        return 7
    elif Precinct==610:
        return 8
    elif Precinct==620:
        return 9
    elif Precinct==630:
        return 10
    elif Precinct==640:
        return 11           
    elif Precinct==650:
        return 12
    elif Precinct==660:
        return 13
    elif Precinct==670:
        return 14
    elif Precinct==675:
        return 15
    elif Precinct==678:
        return 16
    else:
        return 0    

b['Code']=b['PrecinctCode'].apply(district) 

def plot_Map(Code=1):
    #Set up Variable Zoom Locations
    
    #setup
    if Code==0:
        B= b
    else: 
        B= b[(b['Code'] == Code )] 

        
    data = [[0, [44.948530, -93.096058]], [1, [44.944077, -93.156801]], [2, [44.948579, -93.156889]], [560, [44.961943, -93.136571]],[570, [44.957778, -93.136491]],[580, [44.953957, -93.136486]],[590, [44.948546, -93.136444]],[600, [44.943432, -93.136455]],[610, [44.948430, -93.116693]],[620, [44.959384, -93.116180]],[630, [44.967499, -93.103823]],[640, [44.953985, -93.116155]],[650, [44.945421, -93.114748]],[660, [44.944412, -93.121253]],[670, [44.958913, -93.098777]], [675, [44.967045, -93.116131]],[678, [44.968188, -93.136874]]]
    
    # Create the pandas DataFrame 
    Loc = pd.DataFrame(data, columns = ['Code', 'Location']) 
    #Loc.iloc[Precinct,1:]
    # generate a new map
    FG_map = folium.Map(location=Loc.iloc[Code,1], zoom_start=16,tiles="OpenStreetMap")
    
   

    # for each row in the data, add a cicle marker
    for index, row in B.iterrows(): 
        popup_text = "Address: {}<br> HouseholdSize: {}<br> TotalCount: {}"
        popup_text = popup_text.format(row["Address"], row['TotalMemberHH'], row['Counting'])
        folium.CircleMarker(location=(row["Latitude"],row["Longitude"]),
                            radius=row['Counting']+2,
                            color="#E37222",
                            popup=popup_text,
                            fill=True).add_to(FG_map)
        
    return FG_map

```
